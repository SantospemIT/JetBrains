2. Gráfico de líneas: Tendencias de ventas

Ejercicio: Crea un gráfico de líneas que muestre las ventas mensuales de los 4 productos.

```python

# ventas mensuales
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']  # se crean las listas con ,os nombres de cada mes  y vents mensuales
producto_a = [101, 142, 64, 121, 110]
producto_b = [53, 32, 51, 82, 31]
producto_c = [31, 77, 41, 63, 44]
producto_d = [24, 56, 53, 12, 46]

# Grafico  titulos y etiquetas y funcion del tipo de grafico
plt.figure(figsize=(10, 6))   # nos ayuda a delimitar el tamano del grafico
plt.plot(meses, producto_a, marker='o', label='Producto A')  # se crean las lineas del grafico y funcion define el grafico en linea
plt.plot(meses, producto_b, marker='o', label='Producto B')
plt.plot(meses, producto_c, marker='o', label='Producto C')
plt.plot(meses, producto_d, marker='o', label='Producto D')

# crear los ejes y titulo del grafico
plt.title('Ventas Mensuales de Productos')
plt.xlabel('Mes')
plt.ylabel('Ventas')
plt.legend()


# Mostrar el gráfico
plt.show()
```

3. Gráfico de barras: Comparación de ventas totales

Ejercicio: Compara las ventas totales anuales de los productos en un gráfico de barras.

```python

# Datos de ventas mensuales que quiero graficar
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
producto_a = [101, 142, 64, 121, 110]
producto_b = [53, 32, 51, 82, 31]
producto_c = [31, 77, 41, 63, 44]
producto_d = [24, 56, 53, 12, 46]

# Calcular las ventas totales anuales para cada producto operador matematico
total_a = sum(producto_a)
total_b = sum(producto_b)
total_c = sum(producto_c)
total_d = sum(producto_d)

# Se crea grafico de barras y listas con los nombres de productos y ventas
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
ventas_totales = [total_a, total_b, total_c, total_d]

plt.bar(productos, ventas_totales, color=['blue', 'orange', 'green', 'red'])  #define el grafico en barras funcion del tipo de grafico

# Añadir títulos y etiquetas
plt.title('Ventas Totales Anuales de Productos')
plt.xlabel('Producto')
plt.ylabel('Ventas Totales')

# Mostrar el gráfico
plt.show()
```

4. Gráfico de pastel: Proporción de ventas totales

Ejercicio: Representa la proporción de ventas totales por producto en un gráfico de pastel.

```python

# Datos de ventas mensuales
meses = ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo']
producto_a = [101, 142, 64, 121, 110]
producto_b = [53, 32, 51, 82, 31]
producto_c = [31, 77, 41, 63, 44]
producto_d = [24, 56, 53, 12, 46]

# Calcular las ventas totales anuales para cada producto
total_a = sum(producto_a)
total_b = sum(producto_b)
total_c = sum(producto_c)
total_d = sum(producto_d)

# Crear el gráfico de pastel
productos = ['Producto A', 'Producto B', 'Producto C', 'Producto D']
ventas_totales = [total_a, total_b, total_c, total_d]

plt.pie(ventas_totales, labels=productos, autopct='%1.1f%%', colors=['blue', 'orange', 'green', 'red']) #funcion del tipo de grafico

# Añadir título
plt.title('Proporción de Ventas Totales de Productos')

# Mostrar el gráfico
plt.show()
```

5. Gráfico de dispersión: Relación entre productos

Ejercicio: Analiza la relación entre las ventas de dos productos, por ejemplo, Producto A y Producto B.
